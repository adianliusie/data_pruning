```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter

import numpy as np
import seaborn as sns

x = [0.01, 0.05, 0.1, 0.5, 1.0]
baseline_means = [55.01, 58.48, 75.90, 90.11, 91.51]
baseline_stds  = [0.65,  2.53, 2.70,   0.16,  0.26]

no_pretrain_means = [55.75,  55.82, 55.95, 56.18, 55.80]
no_pretrain_stds  = [0.13,  0.03,    0.30,  0.47,  0.00]

low_loss_means = [51.70,  53.29, 64.71, 89.57, 91.51]
low_loss_stds  = [0.18,   1.21,   2.21,  0.49,  0.26]

high_loss_means = [54.16,   53.47, 53.05, 89.04, 91.51]
high_loss_stds  = [0.15,     0.82,  0.88,  0.52,  0.26]

def std_plot(x_axis, means, stds, **kwargs):
    means = np.array(means)
    stds  = np.array(stds)
    
    lower = means - 0.5*stds
    upper = means + 0.5*stds
    
    plt.plot(x, means, **kwargs)
    #plt.fill_between(x_axis, lower, upper, alpha=0.2)

#ax.xaxis.set_major_formatter(FormatStrFormatter('%.2f'))
sns.set_theme()
plt.xscale('log')

std_plot(x, baseline_means, baseline_stds, marker='o')
std_plot(x, low_loss_means, low_loss_stds, marker='o')
std_plot(x, high_loss_means, high_loss_stds, marker='o')
std_plot(x, no_pretrain_means, no_pretrain_stds, marker='o')

plt.xticks([0.01, 0.05, 0.1, 0.5, 1.0],[0.01, 0.05, 0.1, 0.5, 1.0])
plt.legend(['random (BERT)', 'low loss', 'high loss', 'random (no pretrain)'])
```
